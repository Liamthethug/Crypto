```python
def dechiffrer_césarf(message):
    word_list = set(words.words())
    
    for decalage in range(1, 26):
        resultat = ""
        for i in range(len(message)):
            lettre = message[i]

            if lettre.isupper():
                resultat += chr((ord(lettre) - decalage - 65) % 26 + 65)
            elif lettre.islower():
                resultat += chr((ord(lettre) - decalage - 97) % 26 + 97)
            else:
                resultat += lettre

        # Vérifier si au moins un mot est valide
        mots = resultat.split()
        mots_valides = [word.lower() in word_list for word in mots]
        if sum(mots_valides) > len(mots) / 4 :  # Si plus de la moitié des mots sont valides
            print(f"Décalage {decalage}: {resultat}")
            return resultat

    print("Aucun déchiffrement valide trouvé.")
    return None
```

```python
d = ( " la ", " le ", " les ", " un ", " des ", " une ","Bonjour ", "Bonsoir ", "Salut ", "Bienvenue ", "Merci ", "Pardon ", 
    "Excusez-moi ", "D'accord ", "Non ", "Oui ", "Peut-être ", 
    "Le ", "La ", "Les ", "Un ", "Une ", "Des ", "De ", "Du ", "En ",
    "Et ", "Mais ", "Ou ", "Donc ", "Car ", "Ni ", "Si ", "Que ", "Qui ", 
    "Quoi ", "Où ", "Comment ", "Pourquoi ", "Parce ", "Ce ", "Cet ", 
    "Cette ", "Ces ", "Mon ", "Ma ", "Mes ", "Ton ", "Ta ", "Tes ", 
    "Son ", "Sa ", "Ses ", "Notre ", "Nos ", "Votre ", "Vos ", "Leur ", 
    "Leurs ", "Je ", "Tu ", "Il ", "Elle ", "Nous ", "Vous ", "Ils ", "Elles ", 
    "Il ", "Elle ", "Cela ", "Ça ", "On ", "Il y a ", "Y a-t-il ", 
    "Ne ", "Pas ", "Tout ", "Chaque ", "Quelque ", "Quel ", "Quelles ", 
    "Combien ", "Tant ", "Trop ", "Peu ", "Beaucoup ", "Bien joué", "Même ", 
    "Alors ", "Ensuite ", "Puis ", "D'abord ", "Enfin ", "Cependant ", 
    "Néanmoins ", "Toutefois ", "Au ", "À ce moment ", "Aujourd'hui ", 
    "Jamais ", "Toujours ", "Souvent ", "Rarement ", "Parfois ", "Allons ", 
    "Allez ", "Tiens ", "Eh bien ", "Voilà ", "Voyons ", "Hélas ", "Oh ", "Ah ", "Oups ", 
    "Attention ", "Chut ", "Silence ", "Écoutez ", "Regardez ", "Parfait ","Bravo ")

def dechiffrer_Vigenere(encrypted_message, key):
    decrypted_message = []
    key_length = len(key)
    max_unicode = 1114112

    for i, char in enumerate(encrypted_message): #O(l)
        key_char = key[i % key_length]
        decrypted_message.append(chr((ord(char) - ord(key_char)) % max_unicode))

    return ''.join(decrypted_message)

def find_repetitions(encrypted_message): #O(l**2)
    repetitions = {}
    for i in range(len(encrypted_message)): #O(n)
        for j in range(i + 1, len(encrypted_message)): #O(n)
            if encrypted_message[i] == encrypted_message[j]:
                distance = j - i
                if distance not in repetitions:
                    repetitions[distance] = 0
                repetitions[distance] += 1
    return repetitions

def guess_key_length(repetitions):
    sorted_repetitions = sorted(repetitions.items(), key=lambda x: x[1], reverse=True)
    return sorted_repetitions[0][0] if sorted_repetitions else 1

def guess_key(encrypted_message, key_length): #k la longueur de la clé l la longueur du message et n le nombre de caractères différents
    key = []
    max_unicode = 1114112
    for i in range(key_length): #O(k)
        frequencies = {}
        for j in range(i, len(encrypted_message), key_length): #(O(l-i)/k)
            char = encrypted_message[j]
            if char not in frequencies:
                frequencies[char] = 0
            frequencies[char] += 1
        most_frequent_char = max(frequencies, key=frequencies.get) #O(n)
        key.append(chr((ord(most_frequent_char) - ord(' ')) % max_unicode))
    return ''.join(key)

def tenter_dechiffrement_Vigenere(encrypted_message): #O(l**2 + nlog(n) + (n-1)*k + l) => ordre de O(l**2)
    repetitions = find_repetitions(encrypted_message) #O(l**2)
    key_length = guess_key_length(repetitions) #nlog(n)
    key = guess_key(encrypted_message, key_length) #O((n-1)*k + l)
    return dechiffrer_Vigenere(encrypted_message, key), key #O(l)
```
