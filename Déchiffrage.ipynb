```python
def dechiffrer_césarf(message):
    word_list = set(words.words())
    
    for decalage in range(1, 26):
        resultat = ""
        for i in range(len(message)):
            lettre = message[i]

            if lettre.isupper():
                resultat += chr((ord(lettre) - decalage - 65) % 26 + 65)
            elif lettre.islower():
                resultat += chr((ord(lettre) - decalage - 97) % 26 + 97)
            else:
                resultat += lettre

        # Vérifier si au moins un mot est valide
        mots = resultat.split()
        mots_valides = [word.lower() in word_list for word in mots]
        if sum(mots_valides) > len(mots) / 4 :  # Si plus de la moitié des mots sont valides
            print(f"Décalage {decalage}: {resultat}")
            return resultat

    print("Aucun déchiffrement valide trouvé.")
    return None
```

```python
import string

def dechiffrer_scytale(message_chiffre, n_colonnes):
    # Calcul du nombre de lignes
    n_lignes = len(message_chiffre) // n_colonnes
    if len(message_chiffre) % n_colonnes != 0:
        n_lignes += 1  # Ajouter une ligne si nécessaire
        #raise ValueError("Problème de taille")
    # Réorganiser le message dans les colonnes
    matrice = [''] * n_colonnes
    for i in range(len(message_chiffre)):
        col = i % n_colonnes
        matrice[col] += message_chiffre[i]

    # Reconstituer le message déchiffré en lisant les colonnes ligne par ligne
    message_dechiffre = ''.join(matrice)
    return message_dechiffre

def tenter_dechiffrementScytale(message_chiffre):
    for n_colonnes in range(1, len(message_chiffre) + 1):  # Essayer toutes les tailles de colonnes possibles
        message_dechiffre = dechiffrer_scytale(message_chiffre, n_colonnes)
        if message_dechiffre.startswith(d):
            print(message_dechiffre, n_colonnes)
```

```python
def dechiffrer_césard(encrypted_message):
    max_unicode = 1114112
    for key in range(1, max_unicode):
        decrypted_message = ''.join(chr((ord(char) - key) % max_unicode) for char in encrypted_message)
        if messageFr(decrypted_message):
            return decrypted_message, key
    return None, None
```

```python
def dechiffrer_césar4(encrypted_message):
    max_unicode = 1114112  

    def dechiffrer_key(encrypted_message, key_even, key_odd):
        decrypted_message = []
        for i, char in enumerate(encrypted_message):
            if i % 2 == 0:  # Position paire
                decrypted_message.append(chr((ord(char) - key_even) % max_unicode))
            else:  # Position impaire
                decrypted_message.append(chr((ord(char) - key_odd) % max_unicode))
        return ''.join(decrypted_message)

    for key_even in range(1, max_unicode):
        for key_odd in range(1, max_unicode):
            decrypted_message = dechiffrer_key(encrypted_message, key_even, key_odd)
            if messageFr(decrypted_message):
                return decrypted_message, key_even, key_odd
            
    return None, None, None
```

```python
def OddeveenCesar(encrypted_message):  
    repetitions1=[]
    repetitions2=[]
    
    for q, mot in enumerate(encrypted_message):
        dedans=0
        if q % 2 == 0:  # Position paire
           for repet1 in repetitions1: #on regarde si le mot est déjà dans la listes des mots répétés (repetion)...
                if repet1[0]== mot: # ...si ce mot est déjà dans répétions, on augmente son nombre d’apparitions de 1
                    repet1[1]+=1
                    dedans=1
                    break
        else :
            for repet2 in repetitions2:
                if repet2[0]== mot: # ...si ce mot est déjà dans répétions, on augmente son nombre d’apparitions de 1
                    repet2[1]+=1
                    dedans=1
                    break
        if dedans == 0: # ... et s'il n'est pas déjà dedans, on le rajoute et on règle son nombre d'apparitions à 1.
            if q % 2 == 0 :
                repetitions1.append([mot,1])
            else :
                repetitions2.append([mot,1])
                
    repetitions1.sort(key=lambda x: x[1], reverse=True) #on trie les mots par nombre d'apparitions
    repetitions2.sort(key=lambda x: x[1], reverse=True)
    even_most_present = repetitions1[0] #de la forme [charactere, nombre d'apparitions]
    odd_most_present = repetitions2[0] #de la forme [charactere, nombre d'apparitions]
    even_key = ord(even_most_present[0]) - ord(' ') #on récupère la clé de déchiffrement
    odd_key = ord(odd_most_present[0]) - ord(' ')
    message_decrypte = dechiffrer_key(encrypted_message, even_key, odd_key)

    if messageFr(message_decrypte):
        return message_decrypte, (even_key, odd_key)

def dechiffrer_key(encrypted_message, even_key, odd_key):
    max_unicode = 1114112 
    decrypted_message = []
    for i, char in enumerate(encrypted_message):
        if i % 2 == 0:  # Position paire
            decrypted_message.append(chr((ord(char) - even_key) % max_unicode))
        else:  # Position impaire
            decrypted_message.append(chr((ord(char) - odd_key) % max_unicode))
    return ''.join(decrypted_message)
```

```python
def dechiffrer_Vigenere(encrypted_message, key):
    decrypted_message = []
    key_length = len(key)
    max_unicode = 1114112

    for i, char in enumerate(encrypted_message): #O(l)
        key_char = key[i % key_length]
        decrypted_message.append(chr((ord(char) - ord(key_char)) % max_unicode))

    return ''.join(decrypted_message)

def find_repetitions(encrypted_message): #O(l**2)
    repetitions = {}
    for i in range(len(encrypted_message)): #O(n)
        for j in range(i + 1, len(encrypted_message)): #O(n)
            if encrypted_message[i] == encrypted_message[j]:
                distance = j - i
                if distance not in repetitions:
                    repetitions[distance] = 0
                repetitions[distance] += 1
    return repetitions

def guess_key_length(repetitions):
    sorted_repetitions = sorted(repetitions.items(), key=lambda x: x[1], reverse=True)
    return sorted_repetitions[0][0] if sorted_repetitions else 1

def guess_key(encrypted_message, key_length): #k la longueur de la clé l la longueur du message et n le nombre de caractères différents
    key = []
    max_unicode = 1114112
    for i in range(key_length): #O(k)
        frequencies = {}
        for j in range(i, len(encrypted_message), key_length): #(O(l-i)/k)
            char = encrypted_message[j]
            if char not in frequencies:
                frequencies[char] = 0
            frequencies[char] += 1
        most_frequent_char = max(frequencies, key=frequencies.get) #O(n)
        key.append(chr((ord(most_frequent_char) - ord(' ')) % max_unicode))
    return ''.join(key)

def tenter_dechiffrement_Vigenere(encrypted_message): #O(l**2 + nlog(n) + (n-1)*k + l) => ordre de O(l**2)
    repetitions = find_repetitions(encrypted_message) #O(l**2)
    key_length = guess_key_length(repetitions) #nlog(n)
    key = guess_key(encrypted_message, key_length) #O((n-1)*k + l)
    return dechiffrer_Vigenere(encrypted_message, key), key #O(l)
```

```python
import math
from collections import Counter

def dechiffrer_Vigenere2(encrypted_message, key):
    decrypted_message = []
    key_length = len(key)
    max_unicode = 1114112

    for i, char in enumerate(encrypted_message):
        key_char = key[i % key_length]
        decrypted_message.append(chr((ord(char) - ord(key_char)) % max_unicode))
    return ''.join(decrypted_message)

def find_repetitions2(encrypted_message,fd = 9):
    repetitions = {}
    found_distances = 0
    for i in range(len(encrypted_message) - 2):
        seq1 = encrypted_message[i:i+3]
        for j in range(i + 3, len(encrypted_message) - 2):
            seq2 = encrypted_message[j:j+3]
            if seq1 == seq2:
                distance = j - i
                if distance not in repetitions:
                    repetitions[distance] = 0
                repetitions[distance] += 1
                if repetitions[distance] == 2:
                    found_distances += 1
                if found_distances >= fd:
                    #print("repetitions:",repetitions)
                    #print({k: v for k, v in repetitions.items() if v > 1})
                    return {k: v for k, v in repetitions.items() if v > 1}

def guess_key_length2(repetitions):
    if not repetitions:
        return 1
    distances = list(repetitions.keys())
    gcd = distances[0]
    for distance in distances[1:]:
        gcd = math.gcd(gcd, distance)
    #print("gcd:",gcd)
    return gcd

def guess_key2(encrypted_message, key_length):
    key = []
    max_unicode = 1114112
    for i in range(key_length):
        segment = encrypted_message[i::key_length]
        frequencies = Counter(segment)
        most_frequent_char = frequencies.most_common(1)[0][0]
        key.append(chr((ord(most_frequent_char) - ord(' ')) % max_unicode))
    #print("most_frequent_char:",most_frequent_char)
    #print("key:",key)
    return ''.join(key)

def tenter_dechiffrement_Vigenere2(encrypted_message):
    repetitions = find_repetitions2(encrypted_message, 9)
    key_length = guess_key_length2(repetitions)
    key = guess_key2(encrypted_message, key_length)
    return dechiffrer_Vigenere2(encrypted_message, key), key
```

```python
def dechiffrer_enigma(message):
    # Définir les rotors
    rotors = [r1, r2, r3, r4, r5, r6, r7, r8]
    
    decrypted_message = []
    max_unicode = 1114112  # Valeur maximale pour Unicode
    
    for i, char in enumerate(message):
        # Convertir le caractère en entier si nécessaire
        char_code = ord(char) if isinstance(char, str) else char
        
        # Appliquer les rotors dans l'ordre inverse pour déchiffrement
        for rotor in reversed(rotors):
            char_code = rotor.index(char_code)  # Trouver l'indice correspondant dans le rotor
        
        # Calculer le caractère déchiffré
        decrypted_char_code = (char_code - (i % len(rotors))) % max_unicode
        decrypted_char = chr(decrypted_char_code)
        decrypted_message.append(decrypted_char)
    
    return ''.join(decrypted_message)
```
